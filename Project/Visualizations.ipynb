**Visualisatie 1: Taartdiagram (Pie chart)**

Toepassing: Helpt de ongelijkheid visueel snel te begrijpen

Tools: Python met matplotlib


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df1 = pd.read_csv("StudentPerformanceFactors.csv")


# Stel de stijl van de plot in
sns.set(style="whitegrid")

# Maak de boxplot
plt.figure(figsize=(10, 10))
sns.boxplot(x="Internet_Access", y="Exam_Score", data=df1)

# Voeg titels en labels toe
plt.title("Exam Score per Internet Access")
plt.xlabel("Internet Access")
plt.ylabel("Exam Score")

# Toon de plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
labels = ['Yes', 'No']
values = [6108, 499]
colors = ['#66b3ff', '#ff6666']

# Plot
plt.figure(figsize=(6, 4))
bars = plt.bar(labels, values, color=colors)

# Styling
plt.title('Aantal studenten met en zonder internettoegang')
plt.ylabel('Aantal studenten')
plt.xlabel('Internettoegang')

# Waardes op de balken
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 50, int(yval), ha='center', fontweight='bold')

plt.ylim(0, max(values) + 500)
plt.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

# Load your datasets
df1 = pd.read_csv("StudentPerformanceFactors.csv")
df2 = pd.read_csv("gameandgrade.csv")
df3 = pd.read_csv("ResearchInformation3.csv")

# Select dataset with required columns
for df_candidate in [df1, df2, df3]:
    if {"Gaming", "Attendance", "Overall"}.issubset(df_candidate.columns):
        df = df_candidate
        break
else:
    raise ValueError("No dataset contains all required columns.")

# Clean data
df3_clean = df[["Gaming", "Attendance", "Overall"]].dropna().copy()

# Category orders
gaming_order = ["More than 3 Hours", "2-3 Hours", "0-1 Hour"]
attendance_order = ["Below 40%", "40%-59%", "60%-79%", "80%-100%"]
df3_clean["Gaming"] = pd.Categorical(df3_clean["Gaming"], categories=gaming_order, ordered=True)
df3_clean["Attendance"] = pd.Categorical(df3_clean["Attendance"], categories=attendance_order, ordered=True)

# Create one merged trace for "All"
trace_all = go.Scatter(
    x=df3_clean["Attendance"],
    y=df3_clean["Gaming"],
    mode='markers',
    name="All",
    visible=True,  # Default visible
    marker=dict(
        size=df3_clean["Overall"] * 15,
        color=df3_clean["Overall"],
        colorscale="Plasma",
        showscale=True,
        colorbar=dict(title="Overall Grade (CGPA)")
    ),
    text=[f"Attendance: {a}<br>Gaming: {g}<br>CGPA: {c:.2f}" for a, g, c in zip(df3_clean["Attendance"], df3_clean["Gaming"], df3_clean["Overall"])],
    hoverinfo='text'
)

# Create one trace per individual group (initially hidden)
individual_traces = []
for group in gaming_order:
    group_data = df3_clean[df3_clean["Gaming"] == group]
    trace = go.Scatter(
        x=group_data["Attendance"],
        y=group_data["Gaming"],
        mode='markers',
        name=group,
        visible=False,
        marker=dict(
            size=group_data["Overall"] * 15,
            color=group_data["Overall"],
            colorscale="Plasma",
            showscale=True,  # Each individual trace shows its own colorbar
            colorbar=dict(title="Overall Grade (CGPA)")
        ),
        text=[f"Attendance: {a}<br>CGPA: {c:.2f}" for a, c in zip(group_data["Attendance"], group_data["Overall"])],
        hoverinfo='text'
    )
    individual_traces.append(trace)

# Combine all traces
all_traces = [trace_all] + individual_traces

# Dropdown logic: 1 for "All", rest for each group
dropdown_buttons = [
    dict(
        label="All",
        method="update",
        args=[
            {"visible": [True] + [False]*len(gaming_order)},
            {"title": "Gaming Group: All"}
        ]
    )
]

for i, group in enumerate(gaming_order):
    visible_array = [False] * (len(gaming_order) + 1)
    visible_array[i + 1] = True  # Turn on only this group trace
    dropdown_buttons.append(
        dict(
            label=group,
            method="update",
            args=[
                {"visible": visible_array},
                {"title": f"Gaming Group: {group}"}
            ]
        )
    )

# Layout with dropdown in the top-left corner
layout = go.Layout(
    title="Gaming Group: All",
    xaxis=dict(title="Attendance Group"),
    yaxis=dict(
        title="Gaming (hours/day)",
        tickvals=gaming_order,
        categoryorder='array',
        categoryarray=gaming_order[::-1]  # More than 3 Hours at top
    ),
    updatemenus=[
        dict(
            buttons=dropdown_buttons,
            direction="down",
            showactive=True,
            x=0.5,          # move to far left
            xanchor="left",
            y=1.15,       # above the plot
            yanchor="top"
        )
    ],
    margin=dict(t=100),  # Add space for title and dropdown
    height=600
)

# Create and show the figure
fig = go.Figure(data=all_traces, layout=layout)
fig.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

# Load your datasets
df1 = pd.read_csv("StudentPerformanceFactors.csv")
df2 = pd.read_csv("gameandgrade.csv")
df3 = pd.read_csv("ResearchInformation3.csv")

# Select dataset with required columns
for df_candidate in [df1, df2, df3]:
    if {"Gaming", "Attendance", "Overall"}.issubset(df_candidate.columns):
        df = df_candidate
        break
else:
    raise ValueError("No dataset contains all required columns.")

# Clean and prepare the data
df_clean = df[["Gaming", "Attendance", "Overall"]].dropna().copy()

gaming_order = ["More than 3 Hours", "2-3 Hours", "0-1 Hour"]
attendance_order = ["Below 40%", "40%-59%", "60%-79%", "80%-100%"]

df_clean["Gaming"] = pd.Categorical(df_clean["Gaming"], categories=gaming_order, ordered=True)
df_clean["Attendance"] = pd.Categorical(df_clean["Attendance"], categories=attendance_order, ordered=True)

# Map categories to numeric values
df_clean["Gaming_num"] = df_clean["Gaming"].cat.codes
df_clean["Attendance_num"] = df_clean["Attendance"].cat.codes

# Create main trace (all data)
trace_all = go.Scatter3d(
    x=df_clean["Attendance_num"],
    y=df_clean["Gaming_num"],
    z=df_clean["Overall"],
    mode='markers',
    name="All",
    visible=True,
    marker=dict(
        size=df_clean["Overall"] * 5,
        color=df_clean["Overall"],
        colorscale="Plasma",
        colorbar=dict(title="Overall Grade (CGPA)"),
        opacity=0.8
    ),
    text=[f"Attendance: {a}<br>Gaming: {g}<br>CGPA: {c:.2f}" 
          for a, g, c in zip(df_clean["Attendance"], df_clean["Gaming"], df_clean["Overall"])],
    hoverinfo='text'
)

# Individual group traces
group_traces = []
for group in gaming_order:
    group_data = df_clean[df_clean["Gaming"] == group]
    trace = go.Scatter3d(
        x=group_data["Attendance_num"],
        y=group_data["Gaming_num"],
        z=group_data["Overall"],
        mode='markers',
        name=group,
        visible=False,
        marker=dict(
            size=group_data["Overall"] * 5,
            color=group_data["Overall"],
            colorscale="Plasma",
            colorbar=dict(title="Overall Grade (CGPA)"),
            opacity=0.8
        ),
        text=[f"Attendance: {a}<br>CGPA: {c:.2f}" 
              for a, c in zip(group_data["Attendance"], group_data["Overall"])],
        hoverinfo='text'
    )
    group_traces.append(trace)

all_traces = [trace_all] + group_traces

# Dropdown to filter by group
dropdown_buttons = [
    dict(
        label="All",
        method="update",
        args=[
            {"visible": [True] + [False]*len(gaming_order)},
            {"title": "Gaming Group: All"}
        ]
    )
]

for i, group in enumerate(gaming_order):
    visibility = [False] * (len(gaming_order) + 1)
    visibility[i + 1] = True
    dropdown_buttons.append(
        dict(
            label=group,
            method="update",
            args=[
                {"visible": visibility},
                {"title": f"Gaming Group: {group}"}
            ]
        )
    )

# Dropdown for interaction control
movement_buttons = [
    dict(
        label="Free Rotate",
        method="relayout",
        args=[{"scene.dragmode": "orbit"}]
    ),
    dict(
        label="Turntable",
        method="relayout",
        args=[{"scene.dragmode": "turntable"}]
    ),
    dict(
        label="Zoom Only",
        method="relayout",
        args=[{"scene.dragmode": "zoom"}]
    ),
    dict(
        label="Static View",
        method="relayout",
        args=[{"scene.dragmode": False}]
    )
]

# Layout
layout = go.Layout(
    title="Gaming Group: All",
    scene=dict(
        xaxis=dict(
            title="Attendance Group",
            tickvals=list(range(len(attendance_order))),
            ticktext=attendance_order,
            range=[-0.5, 3.5]
        ),
        yaxis=dict(
            title="Gaming (hours/day)",
            tickvals=list(range(len(gaming_order))),
            ticktext=gaming_order[::-1],
            range=[-0.5, 2.5]
        ),
        zaxis=dict(
            title="Overall Grade (CGPA)",
            range=[0, 4.5]
        ),
        camera=dict(
            eye=dict(x=1.4, y=1.4, z=0.9)
        )
    ),
    updatemenus=[
        dict(
            buttons=dropdown_buttons,
            direction="down",
            showactive=True,
            x=0.05,
            xanchor="left",
            y=1.15,
            yanchor="top"
        ),
        dict(
            buttons=movement_buttons,
            direction="right",
            showactive=True,
            x=0.4,
            xanchor="left",
            y=1.15,
            yanchor="top"
        )
    ],
    margin=dict(t=100),
    height=700
)

# Final figure
fig = go.Figure(data=all_traces, layout=layout)
fig.show()
```
