**Dataset 1**

Variable: Internet Access (Toegang tot internet thuis)

Type: Categorical (ja of nee). Deze geeft aan of een student thuis toegang heeft tot internet. Dit is relevant omdat toegang tot internet een belangrijke factor is bij het maken van huiswerk, het opzoeken van informatie en het volgen van online les.



**Visualisatie 1: Taartdiagram (Pie chart)**

**doel:** het visueel tonen van de verdeling tussen studenten met en zonder internettoegang. Deze grafiek laat duidelijk zien dat een grote meerderheid wel toegang heeft, terwijl een kleiner maar significant deel dat niet heeft 

**Beschrijving:**

Waarden: 92.45% “Yes” (6108 studenten), 7.55% “No” (499 studenten)

Vorm: Eenvoudige cirkel met twee segmenten

Toepassing: Helpt de ongelijkheid visueel snel te begrijpen

Tools: Python met matplotlib


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df1 = pd.read_csv("StudentPerformanceFactors.csv")


# Stel de stijl van de plot in
sns.set(style="whitegrid")

# Maak de boxplot
plt.figure(figsize=(10, 10))
sns.boxplot(x="Internet_Access", y="Exam_Score", data=df1)

# Voeg titels en labels toe
plt.title("Exam Score per Internet Access")
plt.xlabel("Internet Access")
plt.ylabel("Exam Score")

# Toon de plot
plt.tight_layout()
plt.show()
```


**Visualisatie 2: Staafdiagram (Bar chart)**
**Doel:**
Het tonen van de absolute aantallen studenten met en zonder internettoegang, om het daadwerkelijke aantal getroffen studenten (499) duidelijk zichtbaar te maken.

**Beschrijving:**

Vorm: Verticale staven naast elkaar met aantallen erboven

Extra: Y-as toont frequentie, X-as toont “Yes” en “No”

Tools: Python met seaborn of matplotlib

```python
import matplotlib.pyplot as plt

# Data
labels = ['Yes', 'No']
values = [6108, 499]
colors = ['#66b3ff', '#ff6666']

# Plot
plt.figure(figsize=(6, 4))
bars = plt.bar(labels, values, color=colors)

# Styling
plt.title('Aantal studenten met en zonder internettoegang')
plt.ylabel('Aantal studenten')
plt.xlabel('Internettoegang')

# Waardes op de balken
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 50, int(yval), ha='center', fontweight='bold')

plt.ylim(0, max(values) + 500)
plt.show()
```

###  Visualisatie 3 – Staafdiagram: ‘Gaming’ (uren per dag)

**Perspectief:**  
Gamen als ontspanning verbetert welzijn en stressmanagement.

**Argument:**  
De meerderheid van de studenten gamet 2 uur of meer per dag, wat suggereert dat gamen een gangbare vorm van ontspanning is.

### Visualisatie 4 – Staafdiagram: ‘Preparation’ (studievoorbereidingstijd)

**Perspectief:**  
Gamen leidt tot slechtere prestaties door minder studietijd.

**Argument:**  
Slechts een klein deel van de studenten bereidt zich meer dan 3 uur per dag voor, wat mogelijk verband houdt met intensief gamen.

###  Visualisatie 6 – Scatterplot: ‘Playing Hours’ vs ‘Grade’

**Perspectief:**  
Gamen leidt tot slechtere prestaties door minder tijd voor studie en slaap.

**Argument:**  
Als er een negatieve relatie zichtbaar is tussen speeltijd en cijfers, ondersteunt dit de stelling dat veel gamen leerprestaties schaadt.

**Beschrijving:**  
Deze scatterplot toont individuele studenten, waarbij op de x-as de gemiddelde gametijd per dag staat, en op de y-as het bijbehorende cijfer. Elk punt representeert een student.

**Toepassing:**  
Deze grafiek helpt bij het herkennen van trends, zoals afnemende cijfers bij hogere gametijd. Als er een negatieve correlatie zichtbaar is, biedt dat ondersteuning voor de kritische kijk op intensief gamen.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Dataset inlezen (indien nog niet gedaan)
df = pd.read_csv("gameandgrade.csv")
df["Grade"] = pd.to_numeric(df["Grade"], errors="coerce")

# Scatterplot
plt.figure(figsize=(6, 4))
sns.scatterplot(x="Playing Hours", y="Grade", data=df, color='orange', edgecolor='w')

plt.title("Scatterplot: Playing Hours vs Grade")
plt.xlabel("Playing Hours per Day")
plt.ylabel("Grade")
plt.grid(True)
plt.tight_layout()
plt.show()
```

###  Visualisatie 7 – Boxplot: ‘Playing Years’ vs ‘Grade’

**Perspectief:**  
Gamen leidt tot slechtere prestaties op lange termijn.

**Argument:**  
Studenten met langdurige game-ervaring kunnen lagere cijfers halen. Als dit zichtbaar wordt in de spreiding per categorie, toont dit een mogelijk cumulatief effect van langdurig gamen.

**Beschrijving:**  
De boxplot laat de spreiding van cijfers zien per ervaringsjaar (0 t/m 4 jaar). Elke box geeft inzicht in mediaan, spreiding en mogelijke uitschieters binnen elke groep.

**Toepassing:**  
Deze visualisatie maakt het mogelijk om prestaties van ervaren en onervaren gamers direct te vergelijken. Grote verschillen in prestaties kunnen een belangrijke bijdrage leveren aan de discussie over de langetermijneffecten van gamen.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Boxplot
plt.figure(figsize=(6, 4))
sns.boxplot(x="Playing Years", y="Grade", hue="Playing Years", data=df,
            palette="Set2", dodge=False, legend=False)

plt.title("Boxplot: Playing Years vs Grade")
plt.xlabel("Playing Years")
plt.ylabel("Grade")
plt.grid(True)
plt.tight_layout()
plt.show()
```
